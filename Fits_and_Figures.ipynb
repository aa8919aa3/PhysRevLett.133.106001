### Function for fitting of Ic- and Ic+ simultaneously - Explicit model

```python
# CPR_2D_model calculates both Ic- and Ic+ of the full device from equations 3 and 4 and returns it in a numpy (2,:) numpy array

def CPR_2Dmodel(x, **kw):    #x= magnetic field data, kw are additionnal parameters to the function.
    Ic1 = kw.get('Ic1', 1)                      # Critical current coefficient of junction 1 (I' in the paper)
    Ic2 = kw.get('Ic2', 1)                      # Critical current coeffiicent of junction 2 (I' in the paper)
    Ic3 = kw.get('Ic3', 2)                      # Critical current of junction 3
    T1 = kw.get('T1', 1)                        # Transmission of JJ1
    T2 = kw.get('T2', 1)                        # Transmission of JJ2
    phi_mx = kw.get('phi_mx', 0)                # phase of the reference branch CPR maximum
    phi_shift = kw.get('phi_shift',0)           # Free phase offset
    S1 = kw.get('S1', 1)                        # B field to flux1 coefficient (Includes the current to Bfield conversion)
    S2 = kw.get('S2', 1)                        # B field to flux2 coefficient (Includes the current to Bfield conversion)
    L3 = kw.get('L3', 0)                        # Reference branch inductance
    offset = kw.get('offset',0)

    def CPR(phi,Ic,T):                          # single JJ model from equation 4
        return Ic * np.sin(phi) / np.sqrt(1 - T*np.sin(phi/2)**2)
    
    # definition of flux contributions to phases
    fx_0 = 1                                    #the flux quantum fx_0 is set to 1 instead of 2.068e-15
    fx1 = 2*np.pi*x*S1/fx_0
    fx2 = 2*np.pi*x*S2/fx_0
    fxL3 = 2*np.pi*L3*Ic3/fx_0
    
    
    # definition of the contributions to the SQUID current
    I1 = CPR(phi_mx + fx2 + fx1,Ic1,T1)                   # Current in JJ1
    I2 = CPR(phi_mx + fx2,Ic2,T2)                         # Current in JJ2
    
    Imoins= offset - Ic3 + CPR(-phi_shift - phi_mx + fx1 + fx2 - fxL3, Ic1, T1) + CPR(-phi_shift - phi_mx + fx2 - fxL3, Ic2, T2)
    Iplus = offset + Ic3 + CPR(-phi_shift + phi_mx + fx1 + fx2 + fxL3, Ic1, T1) + CPR(-phi_shift + phi_mx + fx2 + fxL3, Ic2, T2)
    
    return np.array([Imoins, Iplus])


def CPR_2DFit(data, axeX, **kw):                           # This is the function performing the fit. data has the shape [Ic-array, Ic+array], axeX is the magnetic field values
    def fcn2min(params, axeX, data, **kw):                 # This is the cost function
        param = {}
        for key, value in params.items():                  # Formating parameter dictionnary for proper input in the model function
            param[key] = value.value
        model = CPR_2Dmodel(axeX,  **param)                # Run the model with given set of parameters
        return np.sum(  (model - data)**2  ,  axis = 0 )   # Sum both Ic+ and Ic- least square residues in a single array to enable single minimization

    params = Parameters()
    
                       # Getting all parameters from the kw input
    params.add('Ic1', value=kw.get('Ic1', 0.1), min=kw.get('Ic1_min', None), max=kw.get('Ic1_max', None), expr=kw.get('Ic1_expr', None), vary=kw.get('Ic1_vary', True))
    params.add('Ic2', value=kw.get('Ic2', 1), min=kw.get('Ic2_min', None), max=kw.get('Ic2_max', None), expr=kw.get('Ic2_expr', None), vary=kw.get('Ic2_vary', True))
    params.add('Ic3', value=kw.get('Ic3', 0.1), min=kw.get('Ic3_min', None), max=kw.get('Ic3_max', None), expr=kw.get('Ic3_expr', None), vary=kw.get('Ic3_vary', True))
    params.add('T1', value=kw.get('T1', 1), min=kw.get('T1_min', None), max=kw.get('T1_max', None), expr=kw.get('T1_expr', None), vary=kw.get('T1_vary', True))
    params.add('T2', value=kw.get('T2', 0.1), min=kw.get('T2_min', None), max=kw.get('T2_max', None), expr=kw.get('T2_expr', None), vary=kw.get('T2_vary', True))
    params.add('phi_mx', value=kw.get('phi_mx', 0), min=kw.get('phi_mx_min', None), max=kw.get('phi_mx_max', None), expr=kw.get('phi_mx_expr', None), vary=kw.get('phi_mx_vary', True))
    params.add('phi_shift', value=kw.get('phi_shift', 0), min=kw.get('phi_shift_min', None), max=kw.get('phi_shift_max', None), expr=kw.get('phi_shift_expr', None), vary=kw.get('phi_shift_vary', True))
    params.add('S1', value=kw.get('S1', 0.1), min=kw.get('S1_min', None), max=kw.get('S1_max', None), expr=kw.get('S1_expr', None), vary=kw.get('S1_vary', True))
    params.add('S2', value=kw.get('S2', 1), min=kw.get('S2_min', None), max=kw.get('S2_max', None), expr=kw.get('S2_expr', None), vary=kw.get('S2_vary', True))
    params.add('L3', value=kw.get('L3', 0.1), min=kw.get('L3_min', None), max=kw.get('L3_max', None), expr=kw.get('L3_expr', None), vary=kw.get('L3_vary', True))
    params.add('offset', value=kw.get('offset', 0.1), min=kw.get('offset_min', None), max=kw.get('offset_max', None), expr=kw.get('offset_expr', None), vary=kw.get('offset_vary', True))


    result = minimize(fcn2min, params, args=(axeX, data))    # Performing cost function minimization versus params                    
    if kw.get("show_report", True):
        report_fit(result, show_correl=False)

    return result
```

### Performing the fit

```python
# Input parameter guesses and constrains - Necessary due to the large number of parameters
d = {}

d['Ic1'] = 83
d['Ic2'] = 101
d['Ic3'] = 1264

d['T1'] = 0.5
d['T2'] = 0.5
d['phi_mx'] = 2.0

d['S1']= 86/Conv
d['S2']= 1706/Conv
d['L3']=0.00000


d['Ic1_min' ] = 70
d['Ic1_max'] = 130
d['Ic2_min' ] = 70
d['Ic2_max'] = 130    
d['Ic3_min' ] = 1000
d['Ic3_max'] = 1300

d['T1_min' ] = 0.1
d['T1_max'] = 0.99
d['T2_min' ] = 0.1
d['T2_max'] = 0.99
d['phi_mx_min' ] = 0
d['phi_mx_max'] = np.pi

d['Ic1_vary'] = True
d['Ic2_vary'] = True
d['Ic3_vary'] = True

d['T1_vary'] = True
d['T2_vary'] = True
d['phi_mx_vary'] = True

d['S1_vary']= True
d['S2_vary']= True
d['L3_vary']= False                         # We won't model L3 here

d['L3_min' ] = 0
d['L3_max'] = 0.0001


d['offset'] = 0
d['offset_min'] =-10
d['offset_max'] = 10
d['offset_vary'] = False                    # No need for offset here

d['phi_shift'] = 0
d['phi_shift_vary'] = True
```

```python
###################### Perform the fit - Yellow curve
ds_fit=ds2b1.loc[{'B_uT':slice(0.004*Conv,0.00698*Conv)}]      # Limit fitting domain due to discontinuity in critical current data

data = np.array([ds_fit['crit_current_neg'].values , ds_fit['crit_current_pos'].values])   # organize data into a (2,:) numpy array to feed the fit function
axeX = ds_fit.coords['B_uT'].values
axeXshort = axeX                                               # Store the smaller magnetic field range for later plot

fit_result = CPR_2DFit(data, axeX,**d)

res2b1 = {}                                  # Storing fit parameter outputs in a dictionnary
for key, value in fit_result.params.items():
    res2b1[key] = value.value

###################### Perform the fit - Red curve
ds_fit=ds2b2

data = np.array([ds_fit['crit_current_neg'].values , ds_fit['crit_current_pos'].values   ])
axeX = ds_fit.coords['B_uT'].values

fit_result = CPR_2DFit(data, axeX,**d)

res2b2 = {}                                  # Storing fit parameter outputs in a dictionnary
for key, value in fit_result.params.items():
    res2b2[key] = value.value
    
###################### Perform the fit - Blue curve
ds_fit=ds2b3

data = np.array([ds_fit['crit_current_neg'].values , ds_fit['crit_current_pos'].values   ])
axeX = ds_fit.coords['B_uT'].values

fit_result = CPR_2DFit(data, axeX,**d)

res2b3 = {}                                  # Storing fit parameter outputs in a dictionnary
for key, value in fit_result.params.items():
    res2b3[key] = value.value

###################### Perform the fit - Yellow curve
ds_fit=ds2b4

data = np.array([ds_fit['crit_current_neg'].values , ds_fit['crit_current_pos'].values   ])
axeX = ds_fit.coords['B_uT'].values

fit_result = CPR_2DFit(data, axeX,**d)

res2b4 = {}                                  # Storing fit parameter outputs in a dictionnary
for key, value in fit_result.params.items():
    res2b4[key] = value.value
```

### Function for fitting of Ic- and Ic+ simultaneously - Expression with maximum value

```python
def CPR(phi,Ic,T):                                                    #Current phase relation of a single junction
        return Ic * np.sin(phi) / np.sqrt(1 - T*np.sin(phi/2)**2)

    
def DSQUID_CPR(phi,flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3,sg=1) :              #Current phase relation of the graph.JJ Double SQUID, sg discriminates negative and positive critical current
    # definition of each junction phases
    fx_0 = 1                                                               #the flux quantum fx_0 is set to 1 instead of 2.068e-15
    fx1 = 2*np.pi*flux*S1/fx_0
    fx2 = 2*np.pi*flux*S2/fx_0 + sg*2*np.pi*L3*Ic3/fx_0                    # L3 contribution is simplified to be given by Ic3 only, enabling explicit definition of Ic
    
    phi1 = phi[:,np.newaxis] + fx1 + fx2
    phi2 = phi[:,np.newaxis] + fx2
    phi3 = phi[:,np.newaxis] + flux*0                                     #flux*0 is used to force proper broadcasting
    
    return CPR(phi1,Ic1,T1) + CPR(phi2, Ic2,T2) + CPR(phi3, Ic3,T3)       # Sum of all junction currents

def DSQUID_Critcurrent_2D(flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3) :           #Compute the Ic+ and Ic- critical currents of the full Double SQUID device from looking at the maximum/minimum of the CPR
    phi_vals = np.concatenate(  ( np.linspace(-np.pi,-np.pi/4,500),np.linspace(np.pi/4,np.pi,500) )  )   #educated guess of for what phase we can find the maximum current
    current_plus =  DSQUID_CPR(phi_vals,flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3,sg=1)
    current_moins = DSQUID_CPR(phi_vals,flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3,sg=-1)
    
    return np.array([np.min(current_moins,0), np.max(current_plus,0)])

# Next function encapsulates the model DSQUID_Critcurrent_2D in a kw environnement compatible with the fitting code. (BB stands for black-box since we numerically find the critical current)
def CPR_BB2Dmodel(x, **kw):    #x=magnetic field data, kw are additionnal parameters to the function.
    Ic1 = kw.get('Ic1', 1)                      # Critical current coefficient of junction 1 (I' in the paper)
    Ic2 = kw.get('Ic2', 1)                      # Critical current coefficient of junction 2 (I' in the paper)
    Ic3 = kw.get('Ic3', 1)                      # Critical current coefficient of junction 3 (I' in the paper)
    T1 = kw.get('T1', 1)                        # Transmission of JJ1
    T2 = kw.get('T2', 1)                        # Transmission of JJ2
    T3 = kw.get('T3', 1)                        # Transmission of JJ3
    S1 = kw.get('S1', 1)                        # B field to flux1 coefficient (Includes the current to Bfield conversion)
    S2 = kw.get('S2', 1)                        # B field to flux2 coefficient (Includes the current to Bfield conversion)
    L3 = kw.get('L3', 1)                        # Reference branch inductance
    phi_shift = kw.get('phi_shift', 1)

    return DSQUID_Critcurrent_2D(x-phi_shift,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3)


def CPR_BB2DFit(data, axeX, **kw):              # Function performing the fit for both Ic- and Ic+ using the black-box model, data has the shape [Ic-array, Ic+array], axeX is the magnetic field values

    def fcn2min(params, axeX, data, **kw):
        param = {}
        for key, value in params.items():       # Formating parameter dictionnary for proper input in the model function
            param[key] = value.value
        model = CPR_BB2Dmodel(axeX,  **param)   # Run the model with given set of parameters
        return np.sum(  (model - data)**2  ,  axis = 0 )   # Sum both Ic+ and Ic- least square residues in a single array to enable single minimization

    params = Parameters()
    
                       # Getting all parameters from the kw input
    params.add('Ic1', value=kw.get('Ic1', 0.1), min=kw.get('Ic1_min', None), max=kw.get('Ic1_max', None), expr=kw.get('Ic1_expr', None), vary=kw.get('Ic1_vary', True))
    params.add('Ic2', value=kw.get('Ic2', 1), min=kw.get('Ic2_min', None), max=kw.get('Ic2_max', None), expr=kw.get('Ic2_expr', None), vary=kw.get('Ic2_vary', True))
    params.add('Ic3', value=kw.get('Ic3', 0.1), min=kw.get('Ic3_min', None), max=kw.get('Ic3_max', None), expr=kw.get('Ic3_expr', None), vary=kw.get('Ic3_vary', True))
    params.add('T1', value=kw.get('T1', 1), min=kw.get('T1_min', None), max=kw.get('T1_max', None), expr=kw.get('T1_expr', None), vary=kw.get('T1_vary', True))
    params.add('T2', value=kw.get('T2', 0.1), min=kw.get('T2_min', None), max=kw.get('T2_max', None), expr=kw.get('T2_expr', None), vary=kw.get('T2_vary', True))
    params.add('T3', value=kw.get('T3', 1), min=kw.get('T3_min', None), max=kw.get('T3_max', None), expr=kw.get('T3_expr', None), vary=kw.get('T3_vary', True))
    params.add('S1', value=kw.get('S1', 0.1), min=kw.get('S1_min', None), max=kw.get('S1_max', None), expr=kw.get('S1_expr', None), vary=kw.get('S1_vary', True))
    params.add('S2', value=kw.get('S2', 1), min=kw.get('S2_min', None), max=kw.get('S2_max', None), expr=kw.get('S2_expr', None), vary=kw.get('S2_vary', True))
    params.add('L3', value=kw.get('L3', 0.1), min=kw.get('L3_min', None), max=kw.get('L3_max', None), expr=kw.get('L3_expr', None), vary=kw.get('L3_vary', True))
    params.add('phi_shift', value=kw.get('phi_shift', 0), min=kw.get('phi_shift_min', None), max=kw.get('phi_shift_max', None), expr=kw.get('phi_shift_expr', None), vary=kw.get('phi_shift_vary', True))
    
    result = minimize(fcn2min, params, args=(axeX, data))    # Performing cost function minimization versus params 
    if kw.get("show_report", True):
        report_fit(result, show_correl=False)

    return result
```

```python
# Definition of the functions giving theoretical phase / currents using parameters from fit results

def DSQUID_Critphase(flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3) :       # Numerically finds the phase of the reference junction at the critical current
    phi_vals = np.linspace(np.pi/4,np.pi,2000)
    current = DSQUID_CPR(phi_vals,flux,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3)
    
    return phi_vals[np.argmax(current,0)]
 
def CPR_BB1Dmodel_phase(x, **kw):    
    Ic1 = kw.get('Ic1', 1)                      # Critical current of junction 1 (only first order term)
    Ic2 = kw.get('Ic2', 1)                      # Critical current of junction 2 (only first order term)
    Ic3 = kw.get('Ic3', 1)                      # Critical current of junction 3
    T1 = kw.get('T1', 1)                        # Transmission of JJ1
    T2 = kw.get('T2', 1)                        # Transmission of JJ2
    T3 = kw.get('T3', 1)                        # Transmission of JJ3
    S1 = kw.get('S1', 1)                        # (Includes the current to Bfield conversion)
    S2 = kw.get('S2', 1)                        # (Includes the current to Bfield conversion)
    L3 = kw.get('L3', 1)                        # Reference branch inductance
    
    # This function encapsulates DSQUID_Critphase in a kw environnement to receive fit results
    return DSQUID_Critphase(x,Ic1,Ic2,Ic3,T1,T2,T3,S1,S2,L3)

def CPR_BB1Dmodel_I3(x, **kw):
    Ic1 = kw.get('Ic1', 1)                      # Critical current of junction 1 (only first order term)
    Ic2 = kw.get('Ic2', 1)                      # Critical current of junction 2 (only first order term)
    Ic3 = kw.get('Ic3', 1)                      # Critical current of junction 3
    T1 = kw.get('T1', 1)                        # Transmission of JJ1
    T2 = kw.get('T2', 1)                        # Transmission of JJ2
    T3 = kw.get('T3', 1)                        # Transmission of JJ3
    S1 = kw.get('S1', 1)                        # (Includes the current to Bfield conversion)
    S2 = kw.get('S2', 1)                        # (Includes the current to Bfield conversion)
    L3 = kw.get('L3', 1)                        # Reference branch inductance
    
    phi = CPR_BB1Dmodel_phase(x, **kw)
    
    # This function computes the  kw environnement to receive fit results
    return CPR(phi,Ic3,T3)
```

## Figure 3.b)

From the fits using the black-box model presented previously, we obtained JJ transmission values of 0.54/0.64 in the strongly N doped regime, 0.57/0.63 in the weakly N doped regime and 0.45/0.48 in the strongly P doped regime, as well as corresponding numerical uncertainties from fits.

To convert this parameter to second harmonic contribution to the critical current "A2/Ic", we first compute the Fourier coefficient $E^n_{energy}$ of the energy phase relation of an SNS Josephson junction $U(\varphi) = E \sqrt{1- sin^2(\varphi /2)}$ using a method similar to Larsen, Thorvald W., et al. "Parity-protected superconductor-semiconductor qubit." Physical review letters 125.5 (2020):056801 (Supplemental Material). We can then deduce the CPR harmonics amplitude coefficients from the energy coefficients as $E^n_{CPR} = nE^n_{energy}$.
The resulting figure "Transparency-to-amplitude_conversion" is provided in the repository.

Convertion was then done manually using this figure

```python
#A2/Ic data in %, for strongly P then weakly N then strongly N doped regimes.
A2_1 = np.array([7.41, 10.62, 10.87])
A2_2 = np.array([6.85, 9.12, 8.60])

#relative uncert.
D_1 = np.array([0.05, 0.065, 0.04])
D_2 = np.array([0.05, 0.1, 0.07])

plt.figure(figsize=(4,4))
X = np.array([-1, 1, 2])

plt.errorbar(X, A2_1, D_1*A2_1,fmt = 'none', capsize = 4, ecolor = 'black',mew=1, elinewidth = 1)
plt.errorbar(X, A2_2, D_2*A2_2,fmt = 'none', capsize = 4, ecolor = 'black',mew=1, elinewidth = 1)
plt.plot(X,A2_1,'s',label = 'JJ1')
plt.plot(X,A2_2,'s',color='crimson',label = 'JJ2')

plt.xlabel('Gate doping vs Dirac point (V)')  
plt.ylabel('$A_2/I_c$ (%)')
plt.xlim([-1.5,2.5])
plt.legend(loc='upper left')
```
